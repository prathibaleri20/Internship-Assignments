```python
y=iris[['sepal.length']]
```

```python
x=iris[['sepal.width']]
```

```python
from sklearn.model_selection import train_test_split;
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
y_train.head()
```

```python
y_test.head()
```

```python
#categorize the data was done till here as train and test data
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
y_test.head()
#actual value
```

```python
y_pred[0:5]
#predicted value
```

```python
#we find out the error in prediction
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,y_pred)
```
