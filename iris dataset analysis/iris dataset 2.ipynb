```python
import pandas as pd
```

```python
iris=pd.read_csv('iris.csv')
```

```python
iris.head()
```

```python
iris[iris['sepal.width']>4]
```

```python
iris[iris['petal.width']>1]
```

```python
iris[iris['petal.width']>2]
```

```python
y=iris[['sepal.length']]
```

```python
x=iris[['sepal.width','petal.length']]
```

```python
from sklearn.model_selection import train_test_split;
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
y_train.head()
```

```python
y_test.head()
```

```python
#categorize the data was done till here as train and test data
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
y_test.head()
#actual value
```

```python
y_pred[0:5]
#predicted value
```

```python
#we find out the error in prediction
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,y_pred)
```
