# covid dataset case study

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import datetime as dt
import numpy as np
```

```python
df=pd.read_csv('covid_19_india.csv',parse_dates=['Date'],dayfirst=True)
```

```python
#looking at earluer dates
df.head()
```

```python
#looking at the latest dates
df.tail()
```

```python
df=df[['Date','State/UT','Cured','Deaths','Confirmed']]
df.columns=['date','state','cured','deaths','confirmed']
```

```python
df.head()
df.tail()
```

```python
today=df[df.date == '2020-04-15']
```

```python
today.head()
```

```python
#sorting data wrt number of vonfirmed cases
max_confirmed_cases=today.sort_values(by='confirmed',ascending=False)
max_confirmed_cases
```

```python
top_states_confirmed=max_confirmed_cases[0:5]
```

```python
top_states_confirmed
```

```python
#marking bar plot for state with top confirmed cases
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x="state",y="confirmed",data=top_states_confirmed,hue="state")
plt.show()
```

```python
#sorting data wrt number of death cases
max_death_cases=today.sort_values(by="deaths",ascending=False)
max_death_cases
```

```python
top_states_death=max_death_cases[0:5]
```

bar plot for states with top death cases

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x="state",y="deaths",data=top_states_death,hue="state")
plt.show()
```
